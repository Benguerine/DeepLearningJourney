## Steps in modelling with tensorflow

```python
# Set random seed
tf.random.set_seed(42)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model.compile(loss= tf.keras.losses.mae, # mea is shortet of mean absulet error
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ["mae"])

# Reshape X to be 2D
X_reshaped = X.reshape(-1, 1)

# Pass the reshaped data (X_reshaped) to the fit method
model.fit(X_reshaped, y, epochs=10)
```

```python
model.predict(np.array([17.0]).reshape(-1, 1))
```

## Improving our model

```python
# 1. Create a model

model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(1)
])


# 2. Compile the model
model.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ["mae"])

# 3. fit the model
model.fit(X_reshaped, y, epochs=100)
```

## **Running experiments to improve a model**

```python
tf.random.set_seed(42)

model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model_1.compile(loss= tf.keras.losses.mae,
                optimizer= tf.keras.optimizers.SGD(),
                metrics=["mae"])

model_1.fit(tf.expand_dims(X_train, axis=-1), y_train, epochs=100)
```

```python
# Make and plot predections for model_1
y_pred_1 = model_1.predict(X_test)
plot_predictions(X_train, y_train, X_test, y_test, y_pred_1)
```

```python
# Calculte model_1 evaluation metrics
mae_1 = tf.keras.metrics.mae(y_test, tf.squeeze(y_pred_1))
mse_1 = tf.keras.metrics.mse(y_test, tf.squeeze(y_pred_1))
mae_1, mse_1
```

```python
# Set random seed
tf.random.set_seed(42)

# Replicate model_1 and add an extra layer
model_2 = tf.keras.Sequential([
  tf.keras.layers.Dense(1),
  tf.keras.layers.Dense(1) # add a second layer
])

# Compile the model
model_2.compile(loss=tf.keras.losses.mae,
                optimizer=tf.keras.optimizers.SGD(),
                metrics=['mae'])

# Fit the model
model_2.fit(tf.expand_dims(X_train, axis=-1), y_train, epochs=100, verbose=2)
```

```python
# Make and plot predictions for model_2
y_pred_2 = model_2.predict(X_test)
plot_predictions(X_train, y_train, X_test, y_test, y_pred_2)
```

```python
# Calculate model_2 metrics
mae_2 = tf.keras.metrics.mae(y_test, tf.squeeze(y_pred_2))
mse_2 = tf.keras.metrics.mse(y_test, tf.squeeze(y_pred_2))
mae_2, mse_2
```

```python
# Set random seed
tf.random.set_seed(42)

# Replicate model_2
model_3 = tf.keras.Sequential([
  tf.keras.layers.Dense(1),
  tf.keras.layers.Dense(1)
])

# Compile the model
model_3.compile(loss=tf.keras.losses.mae,
                optimizer=tf.keras.optimizers.SGD(),
                metrics=['mae'])

# Fit the model (this time for 500 epochs, not 100)
model_3.fit(tf.expand_dims(X_train, axis=-1), y_train, epochs=500, verbose=0)
```

```python
# Make and plot predictions for model_3
y_pred_3 = model_3.predict(X_test)
plot_predictions(X_train, y_train, X_test, y_test, y_pred_3)
```

```python
# Calculate model_3 metrics
mae_3 = tf.keras.metrics.mae(y_test, tf.squeeze(y_pred_3))
mse_3 = tf.keras.metrics.mse(y_test, tf.squeeze(y_pred_3))
mae_3, mse_3
```

## Steps in modelling with tensorflow

```python
# Set random seed
tf.random.set_seed(42)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model.compile(loss= tf.keras.losses.mae, # mea is shortet of mean absulet error
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ["mae"])

# Reshape X to be 2D
X_reshaped = X.reshape(-1, 1)

# Pass the reshaped data (X_reshaped) to the fit method
model.fit(X_reshaped, y, epochs=10)
```

```python
model.predict(np.array([17.0]).reshape(-1, 1))
```

## Improving our model

```python
# 1. Create a model

model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(1)
])


# 2. Compile the model
model.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ["mae"])

# 3. fit the model
model.fit(X_reshaped, y, epochs=100)
```

```python
model.predict(np.array([17.0]).reshape(-1, 1))
```
